```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv("train.csv")
df.head(10)
```

# Remove Units (extract numbers) (PART B)

```python
def extract_number(value):
    if pd.isnull(value):
        return np.nan
    match = re.search(r"(\d+\.?\d*)", str(value))
    return float(match.group(1)) if match else np.nan

df["Mileage"] = df["Mileage"].apply(extract_number)
df["Engine"] = df["Engine"].apply(extract_number)
df["Power"] = df["Power"].apply(extract_number)
df["New_Price"] = df["New_Price"].apply(extract_number)
df.head()
```

# Handle Missing Values (PART A)

```python
# Show missing values per column percentage
print((df.isnull().mean() * 100).round(2))
```

```python
# Drop New_Price (too many missing, 86% missing)
df.drop(columns=["New_Price"], inplace=True)

# Median imputations, these columns are associated with car model, so median is safest for them.
for col in ["Mileage", "Engine", "Power"]:
    df[col] = df[col].fillna(df[col].median())

# Seats (mode), Using mode is best. Most cars are 5 seaters - mode = 5
df["Seats"] = df["Seats"].fillna(df["Seats"].mode()[0])
```

```python
df.head(10)
```

# Encode Categorical Variables (PART C)

```python
df = pd.get_dummies(df, columns=["Fuel_Type", "Transmission"], drop_first=False)
df.head()
```

# Create Car Age Feature (PART D)

```python
df["Car_Age"] = 2025 - df["Year"]
df.head()
```

# SELECT OPERATION (PART E)

```python
df_select = df[["Location", "Name", "Year", "Kilometers_Driven", "Price"]]
df_select.head()
```

# FILTER: Diesel Cars in Hyderabad

```python
df_filter = df[(df["Fuel_Type_Diesel"] == 1) & (df["Location"] == "Hyderabad")]
df_filter.head()
```

# GROUP BY + SUMMARIZE

```python
df_multi_summary = df.groupby("Location").agg(
    avg_price=("Price", "mean"),
    max_price=("Price", "max"),
    min_price=("Price", "min"),
    total_cars=("Name", "count")

).reset_index()

df_multi_summary
```
